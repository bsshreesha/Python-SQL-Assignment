### Matplotlib Tutorial

## Introduction
Matplotlib is a popular Python library for data visualization. It provides various tools to create static, animated, and interactive plots.

---

## Why Use Matplotlib?
- **Easy to Use**: Simple syntax for generating plots.
- **Customizable**: Modify colors, labels, legends, and more.
- **Supports Multiple Backends**: Works with different output formats like PNG, PDF, and interactive visualizations.
- **Integration**: Works well with NumPy, Pandas, and other scientific computing libraries.

---

## Importing Matplotlib
To use Matplotlib, import the necessary module:
```python
import matplotlib.pyplot as plt
import numpy as np
```

````python
import matplotlib.pyplot as plt
import numpy as np
````

## Basic Line Plot

````python
x = np.linspace(0, 10, 100)
#print(x)
y = np.sin(x)
#print(y)
plt.plot(x, y, label='Sine Wave')
plt.xlabel('X values')
plt.ylabel('Y-axis: sin(x)')
plt.title('Basic Line Plot')
plt.legend()
plt.show()
````

## Scatter Plot

````python
x = np.random.rand(50)
print(x)
y = np.random.rand(50)
print(y)
colors = np.random.rand(50)
sizes = 500 * np.random.rand(50)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='viridis')
plt.colorbar()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show()
````

## Bar Chart

````python
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]

plt.bar(categories, values, color=['blue', 'green', 'red', 'purple'], alpha=0.5,label=categories)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.legend()
plt.title('Bar Chart')
plt.show()
````

## Histogram

````python
data = np.random.randn(1000)

plt.hist(data, bins=30, color='blue', alpha=0.6)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
````

## Pie Chart

````python
labels = ['Apple', 'Banana', 'Cherry', 'Date']
sizes = [30, 20, 35, 15]
colors = ['red', 'yellow', 'pink', 'brown']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140,)
plt.title('Pie Chart')
plt.show()
````

## Subplots (Multiple Plots in One Figure)

````python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Line Plot
axes[0, 0].plot(x, y, 'r')
axes[0, 0].set_title('Line Plot')

# Scatter Plot
axes[0, 1].scatter(x, y, color='g')
axes[0, 1].set_title('Scatter Plot')

# Bar Chart
axes[1, 0].bar(categories, values, color='b')
axes[1, 0].set_title('Bar Chart')

# Histogram
axes[1, 1].hist(data, bins=20, color='purple')
axes[1, 1].set_title('Histogram')

plt.tight_layout()
plt.show()
````

## Customizing Plots

````python
plt.plot(x, y, linestyle='--', color='magenta', linewidth=2, marker='o', markersize=5, markerfacecolor='yellow')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Customized Plot')
plt.grid(False)
plt.show()
````
